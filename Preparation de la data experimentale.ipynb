```python
import pandas as pd
import re

# List of all elements in the periodic table (you might need to add more elements)
periodic_table_elements = ['H', 'He', 'Li', 'Be', 'B', 'C', 'N', 'O', 'F', 'Ne', 'Na', 'Mg', 'Al', 'Si', 'P', 'S', 'Cl', 'Ar', 'K', 'Ca', 'Sc', 'Ti', 'V', 'Cr', 'Mn', 'Fe', 'Ni', 'Co', 'Cu', 'Zn', 'Ga', 'Ge', 'As', 'Se', 'Br', 'Kr', 'Rb', 'Sr', 'Y', 'Zr', 'Nb', 'Mo', 'Tc', 'Ru', 'Rh', 'Pd', 'Ag', 'Cd', 'In', 'Sn', 'Sb', 'Te', 'I', 'Xe', 'Cs', 'Ba', 'La', 'Ce', 'Pr', 'Nd', 'Pm', 'Sm', 'Eu', 'Gd', 'Tb', 'Dy', 'Ho', 'Er', 'Tm', 'Yb', 'Lu', 'Hf', 'Ta', 'W', 'Re', 'Os', 'Ir', 'Pt', 'Au', 'Hg', 'Tl', 'Pb', 'Bi', 'Po', 'At', 'Rn', 'Fr', 'Ra', 'Ac', 'Th', 'Pa', 'U', 'Np', 'Pu', 'Am', 'Cm', 'Bk', 'Cf', 'Es', 'Fm', 'Md', 'No', 'Lr']

# Function to extract elements from the index using regex
def extract_elements(index_string):
    elements = re.findall(r'[A-Z][a-z]*', index_string)  # Using regex to find elements
    return [element for element in elements if element in periodic_table_elements]

# Apply the extract_elements function to each row in the "index" column
df['extracted_elements'] = df['index'].apply(extract_elements)

# Extract all unique elements from the "extracted_elements" column
all_unique_elements_calc = list(set(element for sublist in df['extracted_elements'] for element in sublist))

print(all_unique_elements_calc)
len(all_unique_elements_calc)
```

```python
def get_dict(element):
    import re
    l= {}
    res = re.findall('[A-Z][^A-Z]*', element)
    for element in res:
        if '.' in element :
            l[re.findall("(\d+\.\d+|[A-Za-z]+)", element)[0]] = float(re.findall('(\d+\.\d+|[A-Za-z]+)', element )[1])
        if len(re.findall('(\d+|[A-Za-z]+)', element )) == 2 :
            l[re.findall('(\d+|[A-Za-z]+)', element )[0]] = float(re.findall('(\d+|[A-Za-z]+)', element )[1])
        if len(re.findall('(\d+|[A-Za-z]+)', element )) == 1 :
            l[re.findall('(\d+|[A-Za-z]+)', element )[0]] = 1     
    return l
```
